P1.1 Draw square lattice with your own implementation (not build-in
 in some library)

```python
import matplotlib.pyplot as plt

def draw_grid(N, M):
    """
    Draws a grid with N columns and M rows.
    """

    # Create the vertical edges (lines parallel to the y-axis)
    vertical_edges = [((i, i), (0, M)) for i in range(N+1)]  
    # Create the horizontal edges (lines parallel to the x-axis)
    horizontal_edges = [((0, N), (i, i)) for i in range(M+1)]  
    # Create all vertices as points where the lines intersect (Cartesian product of points)
    vertices = [(i, j) for i in range(N+1) for j in range(M+1)]  
    
    # Create a new figure and axis for the plot
    fig, ax = plt.subplots()
    
    # Plot all the horizontal and vertical edges
    for e in horizontal_edges + vertical_edges: 
        ax.plot(*e, c='k')  # Plot lines (edges)
    
    # Plot all the vertices (intersections of the edges)
    for v in vertices:
        ax.plot(*v, 'p', c='r')  # Plot points (vertices)
    
    # Display the plot
    plt.show()

# Example usage
draw_grid(1, 4)
```

Draw chain with your own implementation (not build-in in some
 library).

```python
def draw_chain(n):
    for i in range(n):
        if i != n-1:
            print("O", "--", end=" ")
        else:
            print("O")
```

P.1.3 Draw complete graph with your own implementation (not
 build-in in some library).

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_random_complete_graph(K=6, edge_color='k', vertex_color='r', vertex_marker='p'):
    # Generate random vertices
    vertices = [(np.random.uniform(), np.random.uniform()) for _ in range(K)]  # Choosing K random points
    
    # Generate all edges (pairwise connections between vertices)
    edges = [(i, j) for i in vertices for j in vertices if i != j]  # All pairs of vertices except self-pairs
    
    # Plot edges
    for e in edges:
        plt.plot(*zip(*e), c=edge_color)  # Plot edges with the given color
    
    # Plot vertices
    for v in vertices:
        plt.plot(*v, vertex_marker, c=vertex_color)  # Plot vertices with the given marker and color
    
    # Step 5: Display the plot
    plt.show()

plot_random_complete_graph(K=5)
```

 P1.4 Draw weighted graph (any, but N > 5). 

```python
import networkx as nx
import random
def generate_random_graph_fixed():
    # Define 6 fixed nodes
    nodes = ['A', 'B', 'C', 'D', 'E', 'F'] 
    
    # Create a random graph
    G = nx.Graph()
    
    # Randomly create edges and assign random weights
    for i in range(random.randint(5, 8)):  # Random number of edges
        node1 = random.choice(nodes)
        node2 = random.choice([n for n in nodes if n != node1])
        weight = random.randint(1, 10)
        G.add_edge(node1, node2, weight=weight)
    
    # Define node colors matching the actual number of nodes in the graph
    node_colors = random.choices(['blue', 'green', 'red', 'yellow', 'orange', 'purple'], k=G.number_of_nodes())
    
    # Create layout
    pos = nx.spring_layout(G)
    
    # Draw the graph
    nx.draw(G, pos, with_labels=True, node_color=node_colors)
    
    # Draw edge weights
    labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
    
    # Show the graph
    plt.show()

# Call the function to generate and display a random graph
generate_random_graph_fixed()
```

P1.5 Draw directed graph (any, but N > 5)

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the directed graph G2
G2 = nx.DiGraph()
G2.add_edge('A', 'B')
G2.add_edge('B', 'D')
G2.add_edge('A', 'C')
G2.add_edge('C', 'D')
G2.add_edge('C', 'E')
G2.add_edge('D', 'E')
G2.add_edge('D', 'F')
G2.add_edge('B', 'C')
G2.add_edge('E', 'C')
G2.add_edge('F', 'H')
G2.add_edge('F', 'C')


# Define new node colors
node_colors = ['blue', 'green', 'red', 'black', 'orange', 'purple', 'pink']

# Generate positions for nodes
pos = nx.spring_layout(G2)

# Draw the graph
nx.draw(G2, pos, with_labels=True, node_color=node_colors, node_size=1000, font_color='white')
plt.show()
```
